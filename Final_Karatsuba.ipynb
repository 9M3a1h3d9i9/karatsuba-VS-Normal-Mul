# import 

```python
import matplotlib.pyplot as plt
import datetime
import random
import time
```

# Karatsuba Recursive Algo 

```python
def karatsuba(x, y):
    if x < 10 or y < 10:
        return x * y
    else:
        n = max(len(str(x)), len(str(y)))
        half = n // 2
        a = x // (10 ** (half))  # left part of x
        b = x % (10 ** (half))  # right part of x
        c = y // (10 ** (half))  # left part of y
        d = y % (10 ** (half))  # right part of y
        ac = karatsuba(a, c)
        bd = karatsuba(b, d)
        ad_plus_bc = karatsuba(a+b, c+d)-ac-bd
        return ac * (10 ** (2 * half)) + (ad_plus_bc * (10 ** half)) + bd
```

# Normal Multiplication 

```python
def Multiplication(x, y):
  return x * y
```

## Run 

```python
l1=list()
for i in range(1000):
  l1.append(i)
# print(l1)

l2=list()
l3=list()
i=0
lk=list()
lm=list()
while (i < len(l1)):
  x = random.randint(0,1000000)
  y = random.randint(0,1000000)

  t1 = time.clock_gettime(0)
  # t1_0 = datetime.datetime.now()
  karatsuba(x, y)
  t2 = time.clock_gettime(0)
  # t2_0 = datetime.datetime.now()

  l2.append(t2-t1)
  # l2.append(t2_0 - t1_0)

  t3 = time.clock_gettime(0)
  # t3_0 = datetime.datetime.now()
  Multiplication(x, y)
  t4 = time.clock_gettime(0)
  # t4_0 = datetime.datetime.now()
  
  l3.append(t4-t3)
  # l3.append(t4_0-t3_0)

  i+=1
  lk.append(karatsuba(x, y))
  lm.append(Multiplication(x, y))
# print(lk)  
print(l2)
print(l3)
# plt.plot(l1,l2)
# plt.gcf().autofmt_xdate()

fig = plt.figure(figsize=(5,4), dpi=100)
axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])

axes.plot(l1, l2, label="Karatsuba")
axes.plot(l1, l3, label="Normal Mul")

axes.legend(loc=0)

plt.show()
```
